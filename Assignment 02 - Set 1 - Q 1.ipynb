```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
x = pd.Series([24.23, 25.53, 25.41, 24.14, 29.62, 28.25, 25.81, 24.39, 40.26, 32.95, 91.36, 25.99, 39.42, 26.71, 35.00])
```

```python
name = ['Allied Signal', 'Bankers Trust', 'General Mills', 'ITT Industries', 'J. P. Morgan & co.', 'Lehman Brothers',
       'Marriott', 'MCI', 'Merrill Lynch', 'Microsoft', 'Morgan Stanley', 'Sun Microsystems', 'Travellers', 'US Airways',
       'Warner - Lambert']
```

```python
# Pie Chart
plt.figure(figsize=(6,8))
plt.pie(x, labels=name, autopct = '%1.0f%%')
plt.show()
```

```python
# Box- Plot for finding Outliars
sns.boxplot(x)
```

```python
# Mean
x.mean()
```

```python
# Variance
x.var()
```

```python
# Standard Deviation
x.std()
```
